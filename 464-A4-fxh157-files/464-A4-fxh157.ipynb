#### 1b. Pitch and tone

```python
times = np.arange(0,2000,0.01)
f = 1
alist = [1,1,3,1,2,1]
phase_list = [0,0,0,0,0,0]
a4.plot_harmonics(t=np.arange(0,20,0.01), g=a4.harmonic, f=f, alist=alist, phase_list=phase_list, title="harmonic cosine")
val = np.array([a4.harmonic(t=t, f=f, alist=alist, phase_list=phase_list) for t in times])
scipy.io.wavfile.write("harmonic_sound1.wav", rate=44100, data=val.astype(np.float32))
IPython.display.Audio("harmonic_sound1.wav")
```
